```python
import sys
sys.path.append("..")

from solutions.leetcode_solutions import Solution, VisualizedSolution
from IPython.display import clear_output, HTML
from rich.console import Console
from rich.text import Text
import time

RED = "\033[91m"
GREEN = "\033[92m"
YELLOW = "\033[93m"
RESET = "\033[0m"
```

```python

console = Console()

def show_window(nums, left, right):
    clear_output(wait=True)
    text = Text()

    for i, x in enumerate(nums):
        if left <= i <= right:
            text.append(f"{x} ", style="bold white on blue")
        else:
            text.append(f"{x} ", style="dim")
    
    console.print(text)
    time.sleep(0.4)

def show_window_html(nums, left, right):
    clear_output(wait=True)

    html = "<div style='font-size:24px; font-family:monospace; display:flex;'>"
    for i, x in enumerate(nums):
        if left <= i <= right:
            bg = "#8DF39A"    # stronger green
            color = "black"
            style = "font-weight:bold;"
        else:
            bg = "#E0E0E0"    # neutral gray
            color = "#333333"
            style = ""

        html += (
            f"<div style='padding:10px 14px; margin:4px; "
            f"border-radius:6px; background:{bg}; color:{color}; {style}'>"
            f"{x}"
            f"</div>"
        )

    html += "</div>"

    display(HTML(html))
    time.sleep(0.4)

def show_window_html_new(nums, left, right):
    clear_output(wait=True)

    row1 = "<div style='font-size:24px; font-family:monospace; display:flex;'>"
    for i, x in enumerate(nums):
        bg = "#8DF39A" if left <= i <= right else "#E0E0E0"
        color = "black" if left <= i <= right else "#333"
        style = "font-weight:bold;" if left <= i <= right else ""

        row1 += (
            f"<div style='padding:10px 14px; margin:4px; border-radius:6px; "
            f"background:{bg}; color:{color}; {style}'>"
            f"{x}</div>"
        )
    row1 += "</div>"

    row2 = "<div style='font-size:18px; font-family:monospace; display:flex;'>"
    for i in range(len(nums)):
        if i == left:
            row2 += "<div style='padding:0 20px; margin:4px;'>L⬆️</div>"
        elif i == right:
            row2 += "<div style='padding:0 20px; margin:4px;'>R⬆️</div>"
        else:
            row2 += "<div style='padding:0 20px; margin:4px;'></div>"
    row2 += "</div>"

    display(HTML(row1 + row2))
    time.sleep(0.4)
```
